```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("star_wars.csv")
df.head()
```

Below is all the set up needed to produce the original graph.

```python
df = df.dropna()
print(df.isnull().sum().sum())
```

```python
df = df.rename(columns={
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'Episode 1',
    'Unnamed: 4': 'Episode 2',
    'Unnamed: 5': 'Episode 3',
    'Unnamed: 6': 'Episode 4',
    'Unnamed: 7': 'Episode 5',
    'Unnamed: 8': 'Episode 6',
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'Episode 1 Ranking',
    'Unnamed: 10': 'Episode 2 Ranking',
    'Unnamed: 11': 'Episode 3 Ranking',
    'Unnamed: 12': 'Episode 4 Ranking',
    'Unnamed: 13': 'Episode 5 Ranking',
    'Unnamed: 14': 'Episode 6 Ranking'
})
df.head(5)
```

```python
seen_movies = ['Episode 1', 'Episode 2', 'Episode 3', 'Episode 4', 'Episode 5', 'Episode 6']
rank_movies = ['Episode 1 Ranking', 'Episode 2 Ranking', 'Episode 3 Ranking', 'Episode 4 Ranking', 'Episode 5 Ranking', 'Episode 6 Ranking']

seen_df = df[seen_movies]
rank_df = df[rank_movies]
```

```python
for col in rank_movies:
    df[col] = pd.to_numeric(df[col], errors='coerce')

rank_df = df[rank_movies]

avg_ranks = rank_df.mean()
```

```python
# This was the original graph.
best_score = avg_ranks.min()
best_movie = avg_ranks[avg_ranks == best_score].index[0]

plt.figure(figsize=(8,5))
sns.barplot(x=avg_ranks.index, y=avg_ranks.values, color="skyblue")

plt.title("Average Movie Ranking (Highlighting Favorite)")
plt.xlabel("Movie (Episode)")
plt.ylabel("Average Rank")

plt.text(
    x=list(avg_ranks.index).index(best_movie),
    y=best_score + 0.1,
    s="Top Movie",
    color='red',
    ha='center'
)

plt.xticks(rotation=30, ha='right')

plt.yticks(range(1, int(max(avg_ranks.values)) + 2))

plt.tight_layout()
plt.show()
```

```python
# This is the updated graph
plt.figure(figsize=(8,5))
sns.barplot(
    x=avg_ranks.index,
    y=avg_ranks.values,
    hue=avg_ranks.index,      
    palette="colorblind",            
)
plt.title("Average Movie Ranking (Accessible Redesign)")
plt.xlabel("Movie (Episode)")
plt.ylabel("Average Rank")

plt.text(
    x=list(avg_ranks.index).index(best_movie),
    y=best_score + 0.15,
    s="Top Movie",
    color='black',
    ha='center',
    fontweight='bold'
)

plt.xticks(rotation=30, ha='right')
plt.yticks(range(1, int(max(avg_ranks.values)) + 2))
plt.tight_layout()
plt.show()
```

In this redisigned Star Wars ranking chart, I wanted to improve the color accessibility by switching from the unform sky blue color and normal red annotation to a colorblind palette that Seaborn has. This new and improved chart makes it more accessible to those who are color blind and it clearly shows perceptual differences for everyone while also still highlighting the top ranked movie with a bold text. This chart shows clarity under deuternaopia and proanopia colorblind filters. This redesign also follows visual hierarchy with drawing attention to the top rated film without overly overwhelming the viewer and also follows the inclusive design principles by combining the different colors, shapes and text.
